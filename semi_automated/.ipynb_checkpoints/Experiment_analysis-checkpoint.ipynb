## Select metadata and spheroid movement columns

# Statistical analysis Aggrewell400

## Analysis of Variance (ANOVA)

```r
df_selection <- df %>% filter(Platform == "Aggrewell400")
fit <- lm(outwell_perc ~ Aspiration*Dispension*Day, data=df_selection) # calculate how the factors correlate to each other
result <- anova(fit)  # perform ANOVA for all factors - replicates are needed to obtain a p-value
result
```

```r
# Save the ANOVA result
write.csv(result,"stella/output/experiment/Aggrewell400_anova.csv")
```

## Determine significance

We set the significance level at p = 0.05, but it can be usefull to visualize other levels of p. 

```r
critical_F_001 <- qf(p = 1-.001, df1 = result[1,1], df2 = result$Df[length(result$Df)])
critical_F_01 <- qf(p = 1-.01, df1 = result[1,1], df2 = result$Df[length(result$Df)])
critical_F_05 <- qf(p = 1-.05, df1 = result[1,1], df2 = result$Df[length(result$Df)])
critical_F_1 <- qf(p = 1-.1, df1 = result[1,1], df2 = result$Df[length(result$Df)])

result <- cbind(result, 'CriticalValue' = qf(p = 1-.05, df1 = result[,1], df2 = result$Df[length(result$Df)]))
result_F_sorted <- result[order(-result$`F value` ),] # sort the result so the most significant effects are on top 
result_F_sorted <- result_F_sorted[-nrow(result_F_sorted),]
```

## Plot effect sizes

```r
p <- ggplot(result_F_sorted, aes(x = reorder(rownames(result_F_sorted), result_F_sorted$`F value`), y = result_F_sorted$`F value`))+ 
  geom_bar(stat = "identity")+
  geom_hline(aes(yintercept = critical_F_05, color = "p = 0.05"), size = 2)+
  # geom_hline(aes(yintercept = critical_F_1, color ="p = 0.1"), size = 2) +
  scale_color_manual(values = c("p = 0.05" = "red"))+
  # geom_hline(aes(yintercept = critical_F_01, color ="p = 0.01"), linetype = "longdash") +
  # geom_hline(aes(yintercept = critical_F_001, color ="p = 0.001"), linetype = "longdash") +
  labs(title = "", x = " ", y = "Effect size")+
  theme_bw()+
  theme(legend.title =element_blank(),plot.title = element_text(hjust = 0.5), 
        text = element_text(size = 12),
        strip.text = element_text(size = 12), 
        axis.text = element_text(size = 12, color = "black"),
        legend.text = element_text(size = 12, color = "black"),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())+
  # ylim(0, 6)+
  coord_flip()
p

ggsave(filename = "stella/output/experiment/A400_fplot.png", plot = p, dpi = 300, width = 6, height = 4)
```

## Which features are different between a Aggrewell400 and a Aggrewell800
- CellProfiler output: Info at https://cellprofiler-manual.s3.amazonaws.com/CellProfiler-3.0.0/modules/measurement.html
- There is a difference in sample size here as A400 has +-1000 detectable spheroids, while A800 has only 300. Is there a need to do some sampling in A400 to get identical sample sizes? 

```r
# Remove all columns that don't contain data or are constant
cols_to_keep <- c(6,17:28,30:58,60:61,63:64,68:131)
data <- df_full[,cols_to_keep]
# glimpse(data)
```

```r
pvals <- data.frame("Feature"=colnames(data), "col" = 1:length(data[1,]), "pvalue" = 1:length(data[1,])) # initiate
my_vector <- vector("numeric", length = length(data[1,]))
i <- 1
for (i in 2:length(my_vector)) {
  cond1 <- data %>% filter(Platform == "Aggrewell400")
  cond2 <- data %>% filter(Platform == "Aggrewell800")
  my_vector[i] <- t.test(cond1[,i],cond2[,i])$p.value
}

pvals$pvalue <- p.adjust(my_vector, method="BH")
```

```r
df_selection <- pvals[which(pvals$pvalue < 0.05),]%>% arrange(pvalue) %>% head(30)
p <- ggplot(df_selection, aes(x=reorder(Feature, -pvalue), y = -log10(pvalue)))+
        geom_bar(stat = 'identity')+
        coord_flip()+
        ylim(0,0.05)+
        labs(x = "", y = "-log10(p-adj)", title = "A400 vs A800 differences")+
        theme_classic2()+
        theme(plot.title = element_text(hjust = 0.5))
p

ggsave(plot = p, "stella/output/experiment/feature_differences_benj_Hogcorr.png",dpi = 300, width = 8, height = 8)
```

### Create volume distribution table

```r
df_A400 <- df_A400 %>% arrange(Projected_volume_mm3)
df_A400 <- within(df_A400, acc_sum <- cumsum(Projected_volume_mm3))
df_A400$perc_cumvol <- df_A400$acc_sum/sum(df_A400$Projected_volume_mm3)
```

```r
A400_D90 <- df_A400$Projected_volume_mm3[which(abs(df_A400$perc_cumvol-0.90)==min(abs(df_A400$perc_cumvol-0.90)))]
A400_D50 <- df_A400$Projected_volume_mm3[which(abs(df_A400$perc_cumvol-0.50)==min(abs(df_A400$perc_cumvol-0.50)))]
A400_D10 <- df_A400$Projected_volume_mm3[which(abs(df_A400$perc_cumvol-0.10)==min(abs(df_A400$perc_cumvol-0.10)))]
A400_span <- (A400_D90 - A400_D10)/A400_D50
```

```r
df_A800 <- df_A800 %>% arrange(Projected_volume_mm3)
df_A800 <- within(df_A800, acc_sum <- cumsum(Projected_volume_mm3))
df_A800$perc_cumvol <- df_A800$acc_sum/sum(df_A800$Projected_volume_mm3)
```

```r
A800_D90 <- df_A800$Projected_volume_mm3[which(abs(df_A800$perc_cumvol-0.90)==min(abs(df_A800$perc_cumvol-0.90)))]
A800_D50 <- df_A800$Projected_volume_mm3[which(abs(df_A800$perc_cumvol-0.50)==min(abs(df_A800$perc_cumvol-0.50)))]
A800_D10 <- df_A800$Projected_volume_mm3[which(abs(df_A800$perc_cumvol-0.10)==min(abs(df_A800$perc_cumvol-0.10)))]
A800_span <- (A800_D90 - A800_D10)/A800_D50
```

```r
table <- data.frame(matrix(nrow=2,ncol=10))
colnames(table) <- c("Platform", "V_p_spheroid", "D10","D50","D90","Span","Spheroids","µwells","percentage_µwells_detected","V_detected_p_well")
table$Platform <- c("Aggrewell400", "Aggrewell800")
table$V_p_spheroid <- c(mean(df_summary_A400$Projected_volume_mm3),mean(df_summary_A800$Projected_volume_mm3))
table$D10 <-c(A400_D10, A800_D10)
table$D50 <- c(A400_D50, A800_D50)
table$D90 <- c(A400_D90, A800_D90)
table$Span <- c(A400_span, A800_span)
table$Spheroids <- c(mean(df_summary_A400$spheroids),mean(df_summary_A800$spheroids))
table$µwells <- c(mean(df_summary_A400$microwells),mean(df_summary_A800$microwells))
table$percentage_µwells_detected <- ifelse(table$Platform == "Aggrewell400", 100*table$µwells/1200,100*table$µwells/300)
table$V_detected_p_well <- c(volume_per_well_400,volume_per_well_800)
table$V_p_well <- 100*table$V_detected_p_well/table$percentage_µwells_detected
View(table)     

write.csv(table, "stella/output/experiment/A400vsA800_volumedistribution.csv")
```
