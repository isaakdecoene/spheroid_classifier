# Filter selection for A400 and A800 based on manually segmented images

## Packages 

```r
require("ggplot2")
require("dplyr")
getwd()
```

## Data

```r
dft <- read.csv("stella/input/STELLA_sA400_fullyauto_Merge_50-20_C04_D7/cellprofiler/cp_output_spheroids.csv")
dft$type <- "Training"
dfv <- read.csv("stella/input/STELLA_sA400_fullyauto_Merge_50-20_D06_D7/cellprofiler/cp_output_spheroids.csv")
dfv$type = "Validation"
df <- rbind(dft,dfv)
glimpse(df)
```

## Visualize detected spheroids

### Overlay

```r
# Find centers
center_t <- df %>% 
                filter(type == "Training") %>% 
                select(AreaShape_Center_X,AreaShape_Center_Y) %>%
                colMeans()
center_v <- df %>% 
                filter(type == "Validation") %>% 
                select(AreaShape_Center_X,AreaShape_Center_Y) %>%
                colMeans()
```

```r
p <- ggplot(df, aes(x=AreaShape_Center_X, y=AreaShape_Center_Y, color = type))+
        geom_point(aes(size=AreaShape_EquivalentDiameter))+
        geom_point(x=center_t[1], y=center_t[2], shape = 8, size = 5, color = "Red")+
        geom_point(x=center_v[1], y=center_v[2], shape = 8, size = 5, color = "Blue")+
        theme_classic()
p
```

### Overlay corrected locations

Here we see the Ilastik training has to be improved. Some detected 'objects' beyond the well create a deviation of the mean location. 

```r
df$X_corr <- ifelse(df$type == "Training",df$AreaShape_Center_X - center_t[1], ifelse(df$type == "Validation",df$AreaShape_Center_X - center_v[1],NA))
df$Y_corr <- ifelse(df$type == "Training",df$AreaShape_Center_Y - center_t[2], ifelse(df$type == "Validation",df$AreaShape_Center_Y - center_v[2],NA))

# Find new centers
new_center_t <- df %>% 
                filter(type == "Training") %>% 
                select(X_corr,Y_corr) %>%
                colMeans()
new_center_v <- df %>% 
                filter(type == "Validation") %>% 
                select(X_corr, Y_corr) %>%
                colMeans()

p <- ggplot(df, aes(x=X_corr, y=Y_corr, color = type))+
        geom_point(aes(size=AreaShape_EquivalentDiameter))+
        geom_point(x=new_center_t[1], y=new_center_t[2], shape = 8, size = 5, color = "Red")+
        geom_point(x=new_center_v[1], y=new_center_v[2], shape = 8, size = 5, color = "Blue")+
        theme_classic()
p
```

## Selection criteria for A400 based on histograms

### Filter 1: Compactness < 3

Compactness: (2D only) The mean squared distance of the object’s pixels
from the centroid divided by the area. A filled circle will have a compactness
of 1,with irregular objects or objects with holes having a value greater than 1.

```r
df_selection <- df %>% arrange(-AreaShape_Compactness)
summary(df_selection$AreaShape_Compactness)
hist(df_selection$AreaShape_Compactness, breaks = 50)
```

### Filter 2: Eccentricity < 0.8

Eccentricity: (2D only) The eccentricity of the ellipse that has the same second-moments as the region. The eccentricity is the ratio of the distance between the foci of the ellipse and its major axis length. The value is between 0 and 1. (0 and 1 are degenerate cases; an ellipse whose eccentricity is 0 is actually a circle, while an ellipse whose eccentricity is 1 is a line segment.)

```r
df_selection <- df %>% arrange(AreaShape_Eccentricity)
summary(df_selection$AreaShape_Eccentricity)
hist(df_selection$AreaShape_Eccentricity, breaks = 50)
```

### Filter 1: Area > 500 pxl

Area: (2D only) The number of pixels in the region.

```r
df_selection <- df %>% arrange(-AreaShape_Area)
summary(df_selection$AreaShape_Area)
hist(df_selection$AreaShape_Area, breaks = 100)
```

### Option: Perimeter

Perimeter: (2D only) The total number of pixels around the boundary of each region in the image.

```r
df_selection <- df
summary(df_selection$AreaShape_Perimeter)
hist(df_selection$AreaShape_Perimeter, breaks = 50)
```

### Option: FormFactor

FormFactor: (2D only) Calculated as 4*π*Area/Perimeter2. Equals 1 for a perfectly circular object.

```r
df_selection <- df %>% arrange(-AreaShape_FormFactor) %>% head(350)
summary(df_selection$AreaShape_FormFactor)
hist(df_selection$AreaShape_FormFactor, breaks = 50)
```

### Option: MajorAxisLength

FormFactor: (2D only) Calculated as 4*π*Area/Perimeter2. Equals 1 for a perfectly circular object.

```r
df_selection <- df %>% arrange(-AreaShape_MajorAxisLength)
summary(df_selection$AreaShape_MajorAxisLength)
hist(df_selection$AreaShape_MajorAxisLength, breaks = 50)
```
