## Get experimental metadata from image filename
In the CellProfiler output .csv, the data from all images are stored sequentially with an 'ImageNumber' as well as their Filename. All experimental conditions are noted in the name in the same order, so we can use this to generate some metadata for later selection etc.


```r
metadata <- cbind(df[,c(1,4)], as.data.frame(str_split_fixed(df$FileName_orig, "_", 7))) %>%
    setNames(., c("ImageNumber","Filename_orig","Experiment", "Platform", "Handling", "Merge", "asp_disp", "Well", "Day")) %>%
    select(-Merge)%>%
    separate(col = asp_disp, into = c("Aspiration", "Dispension"), sep = "-", remove = TRUE) %>%
    mutate_all(~gsub(".tif", "", .))%>%
    mutate(across(ImageNumber, factor))
glimpse(metadata)
```

## Let's have a look at our spheroids locations
- The microscope does not seem to be ideally tuned. For location-analysis, we can maybe correct for that later on. 
- When we run this analysis for the full experiment (=24 wells), we should be able to define an area and thus a deviation from area around the medium change location. 

```r
A800_middle <- c(mean(data[which(data$ImageNumber==2),3]), mean(data[which(data$ImageNumber==2),4]))
A400_middle <- c(mean(data[which(data$ImageNumber==1),3]), mean(data[which(data$ImageNumber==1),4]))

p <- ggplot(data, aes(x=Location_Center_X, y = Location_Center_Y, color = ImageNumber))+
        geom_point(aes(size = AreaShape_Area), alpha = 0.5)+
        geom_point(x=A800_middle[1], y=A800_middle[2], color = "Red", size = 5, shape = 8)+
        geom_point(x=A400_middle[1], y=A400_middle[2], color = "Red", size = 5, shape = 8)+
        #facet_wrap(.~ImageNumber)+
        labs(x = "X", y="Y")+
        theme_minimal()
p
```

## Which features are different between a Aggrewell400 and a Aggrewell800
- CellProfiler output: Info at https://cellprofiler-manual.s3.amazonaws.com/CellProfiler-3.0.0/modules/measurement.html
- Now I have 1 A400 and 1 A800 well, so let's try a t-test. This can easily be converted in any other test. 
- There is a difference in sample size here as A400 has +-1000 detectable spheroids, while A800 has only 300. Is there a need to do some sampling in A400 to get identical sample sizes? 

```r
pvals <- data.frame("Feature"=colnames(data), "col" = 1:length(data[1,]), "pvalue" = 1:length(data[1,])) # initiate
my_vector <- vector("numeric", length = length(data[1,]))
i <- 1
for (i in 2:length(my_vector)) {
  cond1 <- data %>% filter(ImageNumber == 1)
  cond2 <- data %>% filter(ImageNumber == 2)
  my_vector[i] <- t.test(cond1[,i],cond2[,i])$p.value
}

pvals$pvalue <- my_vector
p <- ggplot(pvals[which(pvals$pvalue < 0.05),], aes(x=reorder(Feature, -pvalue), y = pvalue))+
        geom_bar(stat = 'identity')+
        coord_flip()+
        labs(x = "", y = "p-value", title = "Feature selection")+
        theme_classic2()+
        theme(plot.title = element_text(hjust = 0.5))
p
# glimpse(pvals)
```
