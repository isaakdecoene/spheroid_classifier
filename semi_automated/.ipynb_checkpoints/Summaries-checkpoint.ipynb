```r
# pxl_per_um = 4600pxl/(36µwells*400um/uwell)
pxl_per_um <- 4600/(36*400)
pxl_per_um
um_per_pxl <- 1/pxl_per_um
um_per_pxl
```

```r
df <- rbind(df_A400, df_A800) %>% 
        mutate("ID" = 1:(length(df_A400[,1])+length(df_A800[,1]))) %>% 
        relocate(ID, .before = ImageNumber) %>%
        relocate(type,.after = ID) %>%
        relocate(image_id, .after = ID)
df <- df %>% 
    mutate(AreaShape_EquivalentDiameter = AreaShape_EquivalentDiameter*um_per_pxl) %>%
    mutate("Projected_volume" = 4/3*pi*(AreaShape_EquivalentDiameter/2)^3)
# glimpse(df)
summary(df$image_id)
```

### Metadata from image filename

```r
metadata <- cbind(df[,c(1,2,3,6)], as.data.frame(str_split_fixed(df$FileName_original, "_", 7))) %>%
    setNames(., c("ID","image_id","type","FileName_original","Experiment", "Platform", "Handling", "Merge", "asp_disp", "Well", "Day")) %>%
    select(-Merge, -type)%>%
    separate(col = asp_disp, into = c("Aspiration", "Dispension"), sep = "-", remove = TRUE) %>%
    mutate_all(~gsub(".tif", "", .)) %>%
    mutate(across(where(is.character), as.factor))%>%
    mutate(ID = 1:length(df[,1]))%>%
    mutate(ID = as.numeric(ID))
glimpse(metadata)
```

## Size distribution
see also: https://www.materials-talks.com/d90-d50-d10-and-span-for-dls/
The parameter D90 should more correctly be labeled as Dv(90). It signifies the point in the size distribution, up to and including which, 90% of the total volume of material in the sample is ‘contained’.  For example, if the D90 is 844nm, this means that 90% of the sample has a size of 844nm or smaller

```r
df_A400 <- df_A400 %>% arrange(Projected_volume)
df_A400 <- within(df_A400, acc_sum <- cumsum(Projected_volume))
df_A400$perc_cumvol <- df_A400$acc_sum/sum(df_A400$Projected_volume)
```

### Define polygon object

Determine a set of coordinates for the vertices. Note that the first and last vertex need to be the same to 'close' the polygon.

https://mhallwor.github.io/_pages/basics_SpatialPolygons

```r
grid <- grid_A400
polist <- list()

for (i in 1:length(grid[,1])) {
    name <- as.character(i)
    square <- as.numeric(grid[i,])
    x_coords <- square[c(5,7,9,11,5)]
    y_coords <- square[c(6,8,10,12,6)]
    poly <- Polygon(cbind(x_coords,y_coords)) # one polygon is created 
    polyclass <- Polygons(list(poly), ID = name) # Give object class 'polygon'
    polist <- c(polist,polyclass)
}

# Turn the individual polygons in a list of spatial polygons
all_polygons <- SpatialPolygons(polist)
```

```r
# If you want to save the polygons in a polygon file...
# library(raster)
# shapefile(x = all_polygons, file = "path/to/output/file.shp")
```

### Get spheroid locations

```r
locations <- dft %>% select(AreaShape_Center_X,AreaShape_Center_Y) %>% SpatialPoints
glimpse(locations)
```

### Overlay points and polygons

there are 2 ways: via fs package or via sp package. The sf seems easier but i cannot get it installed... https://r-spatial.github.io/sf/

So we use the 'sp' package again
https://mhallwor.github.io/_pages/basics_SpatialPolygons
https://gis.stackexchange.com/questions/63793/how-to-overlay-a-polygon-over-spatialpointsdataframe-and-preserving-the-spdf-dat

```r
# Create vector with the polygon each spheroid is a part of. 
overlap <- over(locations , all_polygons , fn = NULL)
# make this into a dataframe
df <- data.frame("spheroid" = 1:length(locations), "microwell" = overlap)
glimpse(df)
# colnames(overlap) <- c("spheroid","microwell")
# overlap
```

### Calculate percentage of spheroids in detectable microwells

```r
percentage <- 100*sum(!is.na(df$microwell))/length(all_polygons)
percentage
```

```r
defects <- 100-percentage
defects
```
